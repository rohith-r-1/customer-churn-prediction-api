```python
import pickle
import numpy as np
import pandas as pd

with open('../data/processed_data.pkl', 'rb') as f:
    data = pickle.load(f)

X_test = data['X_test']
y_test = data['y_test']

with open('../models/best_model.pkl', 'rb') as f:
    best_model = pickle.load(f)

y_proba = best_model.predict_proba(X_test)[:, 1]
```

```python
# Business assumptions (example values)
C_FP = 200   # cost of wrongly targeting a non-churner
C_FN = 5000  # cost of losing a churner
C_TP = 500   # cost of targeting a churner (retention offer)

def business_cost(y_true, y_pred):
    y_true = np.array(y_true)
    y_pred = np.array(y_pred)

    tp = np.sum((y_true == 1) & (y_pred == 1))
    fp = np.sum((y_true == 0) & (y_pred == 1))
    fn = np.sum((y_true == 1) & (y_pred == 0))

    total_cost = tp * C_TP + fp * C_FP + fn * C_FN
    return total_cost, tp, fp, fn
```

```python
from sklearn.metrics import f1_score, recall_score, precision_score

thresholds = np.linspace(0.1, 0.9, 17)  # 0.10, 0.15, ..., 0.90
rows = []

for thr in thresholds:
    y_pred_thr = (y_proba >= thr).astype(int)
    total_cost, tp, fp, fn = business_cost(y_test, y_pred_thr)

    rows.append({
        'threshold': thr,
        'total_cost': total_cost,
        'TP': tp,
        'FP': fp,
        'FN': fn,
        'precision': precision_score(y_test, y_pred_thr, zero_division=0),
        'recall': recall_score(y_test, y_pred_thr),
        'f1': f1_score(y_test, y_pred_thr)
    })

thr_df = pd.DataFrame(rows)
thr_df_sorted = thr_df.sort_values('total_cost')
thr_df_sorted.head(10)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
plt.plot(thr_df['threshold'], thr_df['total_cost'], marker='o')
plt.xlabel('Decision Threshold')
plt.ylabel('Total Cost (lower is better)')
plt.title('Business Cost vs Decision Threshold')
plt.grid(True)
plt.savefig('../plots/11_cost_vs_threshold.png', dpi=150, bbox_inches='tight')
plt.show()
```

```python
best_row = thr_df_sorted.iloc[0]
best_threshold = best_row['threshold']
best_threshold, best_row['total_cost'], best_row['precision'], best_row['recall'], best_row['f1']
```

```python
with open('../models/decision_threshold.txt', 'w') as f:
    f.write(str(best_threshold))
print("Best threshold saved:", best_threshold)
```

```python
from sklearn.metrics import precision_score, recall_score, f1_score

# Default 0.5
y_pred_05 = (y_proba >= 0.5).astype(int)
cost_05, tp_05, fp_05, fn_05 = business_cost(y_test, y_pred_05)

# Best threshold
best_thr = best_threshold
y_pred_best = (y_proba >= best_thr).astype(int)
cost_best, tp_b, fp_b, fn_b = business_cost(y_test, y_pred_best)

print("Default 0.5 -> cost:", cost_05, "TP:", tp_05, "FP:", fp_05, "FN:", fn_05)
print("Best", best_thr, "-> cost:", cost_best, "TP:", tp_b, "FP:", fp_b, "FN:", fn_b)

print("\nRecall @0.5:", recall_score(y_test, y_pred_05))
print("Recall @best:", recall_score(y_test, y_pred_best))
```
